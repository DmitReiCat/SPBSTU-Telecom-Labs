Реализуйте алгоритм Быстрого Преобразования Фурье (БПФ), согласно лемме Дэниелсона-Ланцоша - рекурсивный алгоритм ДПФ. Подробности преобразований, алгоритм и самма лемма представлены в методичке.

Произведём импорт и начнём реализовывать алгоритм:

```python
import numpy as np
PI2 = 2 * np.pi
```

```python
def dft(ys):
    N = len(ys)
    ts = np.arange(N) / N
    freqs = np.arange(N)
    args = np.outer(ts, freqs)
    M = np.exp(1j * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps

def my_fft(ys):
    He = dft(ys[::2])
    Ho = dft(ys[1::2])
    ns = np.arange(len(ys))
    W = np.exp(-1j * PI2 * ns / len(ys))
    return np.tile(He, 2) + W * np.tile(Ho, 2)

def my_recursive_fft(ys):
    if len(ys) == 1:
        return ys
    He = my_fft(ys[::2])
    Ho = my_fft(ys[1::2])
    ns = np.arange(len(ys))
    W = np.exp(-1j * PI2 * ns / len(ys))
    return np.tile(He, 2) + W * np.tile(Ho, 2)
```

Для примера возьмём простой массив:

```python
ys = [0.2,-0.5,-0.6, 0.4]
hs = np.fft.fft(ys)
hs
```
